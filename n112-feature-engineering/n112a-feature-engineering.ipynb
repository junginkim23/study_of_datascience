```
# url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kt%26g/kt%26g_0.csv'
url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/kt%26g/kt%26g.csv'
headers = ['분기', '매출액', '영업이익', '영업이익(발표기준)', '세전계속사업이익',
           '당기순이익', '당기순이익(지배)', '당기순이익(비지배)', '자산총계', '부채총계',
           '자본총계', '자본총계(지배)', '자본총계(비지배)', '자본금', '영업활동현금흐름',
           '투자활동현금흐름', '재무활동현금흐름', '영업이익률', '순이익률', 'ROE(%)',
           'ROA(%)', '부채비율', '자본유보율', 'EPS(원)', 'PER(배)']
```

```
import pandas as pd

df = pd.read_csv(url)
df.columns = headers

#문제1 - 각 분기별 영업이익률을 직접 계산! 
#영업 이익이란? : 해당 회사의 본업에서 얼마나 꾸준하게 성과가 나오고 있는지를 확인해볼 수 있는 지표 
#영업 이익률이란? : 성과를 확인할 수 있는 마진율
#계산: 영업이익 = 매출액 - 매출원가 - 영업비용
#계산: 영업이익률 = 영업이익 / 매출액 * 100 

#1단계: 영업이익과 매출액의 타입 확인
# df.info() 

#[object -> int] 함수
def to_Int(x):
  return int(x.replace(',',''))

#2단계: 영업이익과 매출액 타입 int형으로 변환
df['매출액'] = df['매출액'].apply(to_Int)
df['영업이익'] = df['영업이익'].apply(to_Int)

#3단계: 계산! 
df['영업이익률2'] = (df['영업이익'] / df['매출액']) * 100

#계산 결과 소수점이 무자비하게 많이 나오는 것을 확인! -> round method를 활용해 소수점 2째자리로 변경해줌. 아래는 예시
# df_a의 모든 숫자들을 소수점 2자리까지 반올림하기
# df_a.round(2)
# df_a의 A컬럼은 소수점 1자리까지, C컬럼은 소수점 2자리까지 반올림하기
# df_a.round({'A': 1, 'C': 2})

# df.round({'영업이익률2': 2})

df
```

```
df = pd.read_csv(url)
df.columns = headers
def new_to_int(x):
  index = x.find(',')
  return int(x[:index] + x[index+1:])

#19년도 2분기의 매출액만 바꿀 경우
df[df['분기'] == '19/6']['매출액'].apply(new_to_int)

#데이터 셋의 모든 분기의 매출액 변경
df['매출액']  = df['매출액'].apply(new_to_int)
df
```
